# Fitting camera parameters with satellite image

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

import CameraTransform as ct
```

Then we define the intrinsic parameters of the camera:

```python
# intrinsic camera parameters
f = 6.2
sensor_size = (6.17, 4.55)

# initialize the camera
im = plt.imread("CameraImage.jpg")
cam = ct.Camera(ct.RectilinearProjection(f, sensor=sensor_size, image=im))
```

Then initialize the transformation of the map:

```python
# initialize the map
im_map = plt.imread("SatelliteImage.jpg")
cam_map = ct.MapTransform(im_map, scale=20./57.12)
```

```python
# define the points
marker_cam = np.array([[ 519.273,  730.962,  857.258, 2745.978,  318.579],
                       [2225.455, 1411.264, 1157.528, 1382.157,  971.409]])
marker_map = np.array([[1556.434, 1484.303, 1409.971, 1849.686,  894.264],
                       [2230.033, 1920.227, 1518.964, 1813.804,  617.598]])


# fit the camera parameters form the point correspondences
p = cam.fitCamParametersFromPointCorrespondences(marker_cam, marker_map, cam_map)

# print the fitted camera parameters
print(cam)
```

```python
 # create the figure
plt.figure(0, (8, 3))

# and plot the point correspondences
cam.plotPointCorrespondence(im, im_map, cam_map, marker_cam, marker_map)
```
